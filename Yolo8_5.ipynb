```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
import json
import os
import numpy as np

# Carrega o modelo treinado
model = YOLO("/content/runs/segment/train2/weights/best.pt")

def analyze_defects(image_path, annotations_json=None):
    # Verifica se a imagem existe
    if not os.path.exists(image_path):
        print(f"\nErro: Arquivo de imagem nÃ£o encontrado em {image_path}")
        print("Por favor, verifique o caminho e tente novamente.")
        return

    # Faz a previsÃ£o
    try:
        results = model(image_path)
    except Exception as e:
        print(f"\nErro durante a prediÃ§Ã£o: {e}")
        return

    # Carrega as anotaÃ§Ãµes se o JSON foi fornecido
    ground_truth = None
    image_name = os.path.basename(image_path)

    if annotations_json and os.path.exists(annotations_json):
        try:
            with open(annotations_json, 'r') as f:
                annotations = json.load(f)

                # Procura pelas anotaÃ§Ãµes da imagem atual
                for item in annotations:
                    if item.get('image_key') == image_name:
                        ground_truth = item
                        break
        except Exception as e:
            print(f"\nErro ao ler anotaÃ§Ãµes: {e}")

    # Processa os resultados
    for result in results:
        # Configura a figura
        plt.figure(figsize=(15, 6))

        # Imagem original (com ground truth se disponÃ­vel)
        plt.subplot(1, 2, 1)
        original_img = cv2.cvtColor(result.orig_img, cv2.COLOR_BGR2RGB)

        if ground_truth:
            # Desenha os polÃ­gonos das fissuras
            for crack in ground_truth.get('cracks', []):
                # Converte pontos do polÃ­gono para array numpy
                points = np.array(crack['polygon_points'], np.int32)
                points = points.reshape((-1, 1, 2))

                # Desenha o polÃ­gono
                cv2.polylines(original_img, [points], isClosed=True, color=(0, 255, 0), thickness=2)

                # Desenha a bounding box
                bbox = crack['bounding_box']
                x, y = float(bbox['x']), float(bbox['y'])
                w, h = float(bbox['width']), float(bbox['height'])
                x1, y1 = int(x), int(y)
                x2, y2 = int(x + w), int(y + h)
                cv2.rectangle(original_img, (x1, y1), (x2, y2), (0, 255, 255), 1)

                # Adiciona label
                cv2.putText(original_img, crack['label'], (x1, y1-5),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

            plt.title('Imagem Original (Verde: Ground Truth)')
        else:
            plt.title('Imagem Original')

        plt.imshow(original_img)
        plt.axis('off')

        # Imagem com prediÃ§Ãµes
        plt.subplot(1, 2, 2)
        annotated_img = result.plot()
        plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))

        # Status da anÃ¡lise
        if len(result.boxes) > 0:
            status = "Faulty (Com Defeito)"
            defects_count = len(result.boxes)
            plt.title(f"{status}\n{defects_count} defeito(s) detectado(s)")
        else:
            plt.title("Healthy (SaudÃ¡vel)")

        plt.axis('off')
        plt.tight_layout()
        plt.show()

        # SaÃ­da detalhada
        print("\n=== AnÃ¡lise Detalhada ===")
        print(f"Arquivo analisado: {image_path}")
        print(f"DimensÃµes: {result.orig_img.shape[1]}x{result.orig_img.shape[0]}")

        if len(result.boxes) > 0:
            print("\nDefeitos detectados (Vermelho):")
            for i, box in enumerate(result.boxes, 1):
                xyxy = box.xyxy[0].tolist()
                conf = box.conf.item()
                cls = int(box.cls.item())
                label = model.names[cls]

                print(f"\nDefeito {i}:")
                print(f"- Tipo: {label}")
                print(f"- ConfianÃ§a: {conf:.2%}")
                print(f"- Coordenadas: {[round(x) for x in xyxy]}")
                print(f"- Ãrea: {(xyxy[2]-xyxy[0])*(xyxy[3]-xyxy[1]):.0f} pxÂ²")
        else:
            print("Nenhum defeito detectado.")

        if ground_truth:
            print("\nGround Truth (Verde):")
            print(f"Total de fissuras esperadas: {len(ground_truth.get('cracks', []))}")
            for i, crack in enumerate(ground_truth.get('cracks', []), 1):
                bbox = crack['bounding_box']
                x, y = float(bbox['x']), float(bbox['y'])
                w, h = float(bbox['width']), float(bbox['height'])

                print(f"\nfissura {i}:")
                print(f"- Tipo: {crack['label']}")
                print(f"- Bounding Box: x={x:.1f}, y={y:.1f}, width={w:.1f}, height={h:.1f}")
                print(f"- Ãrea BB: {w*h:.1f} pxÂ²")
                print(f"- Pontos do polÃ­gono: {crack['polygon_points']}")

# Como usar:
# 1. Primeiro liste as imagens disponÃ­veis
train_img_dir = "/content/drive/Shareddrives/Leticia_Vieira_G_Pesquisa/dataset/CAI-SWTB-Dataset/Train/images"
available_images = [f for f in os.listdir(train_img_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

if not available_images:
    print("Nenhuma imagem encontrada no diretÃ³rio especificado.")
else:
    print("Imagens disponÃ­veis para anÃ¡lise:")
    for i, img in enumerate(available_images[:5]):  # Mostra apenas as 5 primeiras
        print(f"{i+1}. {img}")

    # Seleciona a primeira imagem como exemplo
    sample_image = "Areial_Faulty246.jpg"  # Substitua por uma imagem que existe no seu JSON
    image_path = os.path.join(train_img_dir, sample_image)
    annotations_json = "/content/drive/Shareddrives/Leticia_Vieira_G_Pesquisa/dataset/CAI-SWTB-Dataset/Train/cai.json"

    print(f"\nAnalisando imagem: {sample_image}")
    analyze_defects(image_path, annotations_json)
```

```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
import json
import os
import numpy as np

# Carrega o modelo treinado
model = YOLO("/content/runs/segment/train2/weights/best.pt")

def analyze_defects(image_path, annotations_json=None):
    # Verifica se a imagem existe
    if not os.path.exists(image_path):
        print(f"\nErro: Arquivo de imagem nÃ£o encontrado em {image_path}")
        return

    # Faz a previsÃ£o
    try:
        results = model(image_path)
    except Exception as e:
        print(f"\nErro durante a prediÃ§Ã£o: {e}")
        return

    # Carrega as anotaÃ§Ãµes
    ground_truth = None
    image_name = os.path.basename(image_path)

    if annotations_json and os.path.exists(annotations_json):
        try:
            with open(annotations_json, 'r') as f:
                annotations = json.load(f)
                for item in annotations:
                    if item.get('image_key') == image_name:
                        ground_truth = item
                        break
        except Exception as e:
            print(f"\nErro ao ler anotaÃ§Ãµes: {e}")

    # Processa os resultados
    for result in results:
        # Filtra apenas detecÃ§Ãµes de fissuras (ajuste os nomes conforme seu modelo)
        crack_detections = []
        for box in result.boxes:
            cls = int(box.cls.item())
            label = model.names[cls]
            if 'crack' in label.lower():  # Filtra por rÃ³tulos contendo 'crack'
                crack_detections.append(box)

        # Configura a figura
        plt.figure(figsize=(15, 6))

        # Imagem original com ground truth
        plt.subplot(1, 2, 1)
        original_img = cv2.cvtColor(result.orig_img, cv2.COLOR_BGR2RGB)

        if ground_truth:
            for crack in ground_truth.get('cracks', []):
                points = np.array(crack['polygon_points'], np.int32)
                points = points.reshape((-1, 1, 2))
                cv2.polylines(original_img, [points], isClosed=True, color=(0, 255, 0), thickness=2)
                bbox = crack['bounding_box']
                x, y = float(bbox['x']), float(bbox['y'])
                w, h = float(bbox['width']), float(bbox['height'])
                cv2.rectangle(original_img, (int(x), int(y)), (int(x+w), int(y+h)), (0, 255, 255), 1)
                cv2.putText(original_img, crack['label'], (int(x), int(y-5)),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

            plt.title('Ground Truth (Verde)')
        else:
            plt.title('Imagem Original')

        plt.imshow(original_img)
        plt.axis('off')

        # Imagem com prediÃ§Ãµes de fissuras
        plt.subplot(1, 2, 2)
        pred_img = cv2.cvtColor(result.orig_img.copy(), cv2.COLOR_BGR2RGB)

        if len(crack_detections) > 0:
            for box in crack_detections:
                xyxy = [int(x) for x in box.xyxy[0].tolist()]
                cv2.rectangle(pred_img, (xyxy[0], xyxy[1]), (xyxy[2], xyxy[3]), (0, 0, 255), 2)
                conf = box.conf.item()
                label = model.names[int(box.cls.item())]
                cv2.putText(pred_img, f"{label} {conf:.2f}", (xyxy[0], xyxy[1]-10),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

            plt.title(f"PrediÃ§Ãµes (Vermelho)\n{len(crack_detections)} fissura(s) detectada(s)")
        else:
            plt.title("Nenhuma fissura detectada")

        plt.imshow(pred_img)
        plt.axis('off')
        plt.tight_layout()
        plt.show()

        # SaÃ­da detalhada
        print("\n=== AnÃ¡lise Detalhada ===")
        print(f"Arquivo: {image_name}")
        print(f"DimensÃµes: {result.orig_img.shape[1]}x{result.orig_img.shape[0]}")

        # DetecÃ§Ãµes
        if len(crack_detections) > 0:
            print("\nfissuras detectadas:")
            for i, box in enumerate(crack_detections, 1):
                xyxy = [int(x) for x in box.xyxy[0].tolist()]
                conf = box.conf.item()
                label = model.names[int(box.cls.item())]
                area = (xyxy[2]-xyxy[0])*(xyxy[3]-xyxy[1])

                print(f"\nfissura {i}:")
                print(f"- Tipo: {label}")
                print(f"- ConfianÃ§a: {conf:.2%}")
                print(f"- Bounding Box: {xyxy}")
                print(f"- Ãrea: {area} pxÂ²")
        else:
            print("\nNenhuma fissura detectada pelo modelo")

        # Ground Truth
        if ground_truth:
            print("\nGround Truth (fissuras reais):")
            for i, crack in enumerate(ground_truth.get('cracks', []), 1):
                bbox = crack['bounding_box']
                x, y = float(bbox['x']), float(bbox['y'])
                w, h = float(bbox['width']), float(bbox['height'])
                area = w * h

                print(f"\nfissura {i}:")
                print(f"- Tipo: {crack['label']}")
                print(f"- Bounding Box: x={x:.1f}, y={y:.1f}, w={w:.1f}, h={h:.1f}")
                print(f"- Ãrea: {area:.1f} pxÂ²")
                print(f"- Pontos do polÃ­gono: {crack['polygon_points']}")

# Como usar:
train_img_dir = "/content/drive/Shareddrives/Leticia_Vieira_G_Pesquisa/dataset/CAI-SWTB-Dataset/Train/images"
annotations_json = "/content/drive/Shareddrives/Leticia_Vieira_G_Pesquisa/dataset/CAI-SWTB-Dataset/Train/cai.json"

# Analisa uma imagem especÃ­fica (substitua pelo nome correto)
image_name = "Areial_Faulty246.jpg"  # Exemplo que existe no seu JSON
image_path = os.path.join(train_img_dir, image_name)

if os.path.exists(image_path):
    analyze_defects(image_path, annotations_json)
else:
    print(f"Imagem {image_name} nÃ£o encontrada no diretÃ³rio")
```
